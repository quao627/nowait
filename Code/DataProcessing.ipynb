```python
import pandas as pd
import nltk
nltk.download('punkt')
from nltk.tokenize import sent_tokenize
import warnings

warnings.filterwarnings("ignore")
pd.options.display.max_columns = 100
```

```python
reviews = pd.read_csv("../Data/All_reviews_finalized.csv")
restaurants = pd.read_csv("../Data/All_restaurants_finalized.csv")
reviews = reviews.drop(["Unnamed: 0.1", "Unnamed: 0"], axis=1)
```

Average Word Count

```python
word_cnt = reviews["contents"].apply(lambda x: len(x.split(" ")))
reviews["Sent_length"] = word_cnt
avg_word_cnt = word_cnt.mean()
print(avg_word_cnt)
```

Average Sentence Count

```python
sent_cnt = reviews["contents"].apply(lambda x: len(sent_tokenize(x)))
reviews["Sent_length"] = sent_cnt
avg_sent_cnt = sent_cnt.mean()
print(avg_sent_cnt)
```
